```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Graph

```python
x = np.array([0,1,2,3,4])
y = x * 2
y_squared = x ** 2

plt.plot(x, y, label = '2 Multiple')
plt.plot(x, y_squared, label = 'X Squared')
plt.title('Two Multiple Graph', fontdict = {'fontname': 'Sans Serif'})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])

plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (10,6)) 
ax1.plot(x, y, label = 'X * 2', c = 'b', lw = 4, ls = '--')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.legend()

ax2.plot(x, y_squared,
         label = 'X ^ 2',
         c = 'g', marker = '*',
         ms = 10,
         markeredgecolor = 'b')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.legend()

plt.show()
```

```python
fig.savefig('Basic Graph.png')
```

## Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
fig, ax = plt.subplots()
bars = ax.bar(labels, values)

patterns = ['/', 'O', '*']

for x, bar in enumerate(bars):
    bar.set_hatch(patterns[x])

# bars[0].set_hatch('/')
plt.show()
```
